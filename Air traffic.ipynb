```python
pivot=pd.pivot_table(data=df,index='Year',columns='Month',values='PassengerCount',aggfunc='sum')
pivot
```

##### The data shows consistent growth in passenger numbers from 2005 to 2015, particularly in July, August,and December.

##### 2005 and 2016 stands out with many missing or incomplete data points.


```python
plt.figure(figsize=(15,6))
sb.heatmap(pivot,annot=True, cmap='RdBu')
plt.yticks(rotation=90)
plt.show()
```

##### Over the years 2005 to 2016, the number of passengers has increased .

##### The darker the color, the more passengers in that month.July, August, December are the busiest months have the most passengers.

##### February and April have fewer passengers having lighter shades.

##### From 2012 to 2016, passenger numbers increased the most, 2016 had the highest number of passengers.
